## Identifying Outlier and Removing it

```python
# Usual tend is that the price of the house increases with increase in quality
# one point in 9 and 2 point in 10 (lowest point) seems like outlier
# It is supposed to be much more than 200000 for quality 9 and 10.

sns.scatterplot(x='Overall Qual', y='SalePrice', data=df)
```

```python
# for finding those data's which are classified as outlier from the scatterplot
# so condition might be quality might be greater than 8 and sale price less than 200000.

df[(df['Overall Qual'] > 8) & (df['SalePrice'] < 200000)]
```

```python
# another criteria is with increase in liveing area, the sale price also goes high.
# We again see 3 data points in the scatter plot which deviates from this trends.
# this can be classified as outlier.

sns.scatterplot(x='Gr Liv Area', y='SalePrice', data=df)
```

```python
# Again the condition will be for higher living area (4000) and lower sale price (400000).

drop_ind = df[(df['Gr Liv Area'] > 4000) & (df['SalePrice']  < 400000)].index
drop_ind
```

```python
df.head()
```

```python
# Dropping these outlier rows (axis =0) from the original dataset.

df = df.drop(drop_ind, axis=0)
df.head()
```

```python
#Plotting the scatter plot after removing the outlier we identified

sns.scatterplot(x='Gr Liv Area', y='SalePrice', data=df)
```

## Identifying Missing Data

```python
# Converting to percentage, to know what percentage of data are missing.

percentage_missing_data = 100 * (df.isnull().sum())/len(df)
percentage_missing_data
```

```python
# We only need values greater than 0
# sortted the values after finding the missing percentage greater than 0.

percentage_missing_data = percentage_missing_data[percentage_missing_data > 0].sort_values()
percentage_missing_data
```

```python
# Just defeining a function to collabrate the above two steps

def missing_data(dataset):
    percentage_missing_data = 100 * (df.isnull().sum())/len(df)
    percentage_missing_data = percentage_missing_data[percentage_missing_data > 0].sort_values()
    return percentage_missing_data
```

```python
# Plotting the missing value percentage in a scatter plot
# X-axis dertmines the missing value index or the features names
# Y-axis determines their coressponding values

plt.figure(figsize=(12,8))
sns.barplot(x = missing_data(df).index, y = missing_data(df))
plt.xticks(rotation =90);
```

```python
df = df.dropna(axis=0, subset=['Electrical','Garage Area'])
```

```python
# These columns or features have integer values, so we are replacing NaN with 0.
# Since there is no basement it can be 0 for integer columns and None for string columns.

base_num_col = ['Bsmt Unf SF', 'Total Bsmt SF', 'BsmtFin SF 2', 'BsmtFin SF 1', 'Bsmt Full Bath', 'Bsmt Half Bath']
df[base_num_col] = df[base_num_col].fillna(0)
```

```python
base_str_col = ['Bsmt Qual','Bsmt Cond','BsmtFin Type 1','BsmtFin Type 2','Bsmt Exposure']
df[base_str_col] = df[base_str_col].fillna('None')
```

```python
missing_data(df)
```

```python
# we are going to set the y limit between 0 and 1 percent

plt.figure(figsize=(12,8))
sns.barplot(x = missing_data(df).index, y = missing_data(df))
plt.xticks(rotation =90)
plt.ylim(0,1)
```

```python
df['Mas Vnr Type'] = df['Mas Vnr Type'].fillna('None')
df['Mas Vnr Area'] = df['Mas Vnr Area'].fillna(0)
```

```python
# we are going to set the y limit between 0 and 1 percent

plt.figure(figsize=(12,8))
sns.barplot(x = missing_data(df).index, y = missing_data(df))
plt.xticks(rotation =90)
plt.ylim(0,1)
```

```python
# Since we know that all the Garage related rows are string except year_built.
# And there are only very less values missing for this features.
# we will try to fill None for string features and 0 for year built feature.

gar_str_cols = ['Garage Type','Garage Finish','Garage Qual','Garage Cond']
df[gar_str_cols] = df[gar_str_cols].fillna('None')
df['Garage Yr Blt'] = df['Garage Yr Blt'].fillna(0)
```

```python
# About 80 percent of Features Fence, Alley, Misc Feature, Pool QC are NA
# Only Luxury villas might be having the swimming pools, 
# So It is better to remove these features rather than filling it.

df = df.drop(['Fence', 'Alley', 'Misc Feature', 'Pool QC'], axis=1)
df.head()
```

```python
# To see the unique values in 'Fireplace Qu'
# All the values in the columns are string values.

df['Fireplace Qu'].unique()
```

```python
# So we can replace it with "None" for features 'Fireplace Qu'.

df['Fireplace Qu'] = df['Fireplace Qu'].fillna('None')
```

```python
# To see the unique values in 'Lot Frontage'
# All the values in the columns are numeric values.

df['Lot Frontage'].unique()
```

```python
# Since it is numrical feature and there is direct relationship with feature 'neighbourhood'
# So we will group by the column 'Lot Frontage' based on 'Neighborhood' 
# Then replace the NaN values by mean values with help of transform function.


df['Lot Frontage'] = df.groupby('Neighborhood')['Lot Frontage'].transform(lambda value : value.fillna(value.mean()))
```

```python
# still there are 3 null values we can replace it with 0.

df.isnull().sum()
```

```python
# Filling the rest of the features with '0'.

df['Lot Frontage'] = df['Lot Frontage'].fillna(0)
```

```python
# Checking whether there are more null values.

df.isnull().sum()
```

## Dealing with Object Categorical values

```python
# We will be having both string (object) and integer/float datatype in dataset.
# We need to convert the string categorical data to integer values for ML to execute


df.info()
```

```python
# Select the only the object data type

df_object_dtype = df.select_dtypes(include='object')
df_object_dtype
```

```python
# selecting only the integer or float data types

df_int_dtype = df.select_dtypes(exclude='object')
df_int_dtype
```

```python
# Converting categorical string features to the integer values with get_dummies functions.
# Drop_first will remove the first features from the resulting dummies of the corresponding features.

df_object_dummies = pd.get_dummies(df_object_dtype, drop_first=True)
df_object_dummies
```

```python
# now we need to concatenate the converted dummy features and integer features

final_df = pd.concat([df_object_dummies, df_int_dtype], axis=1)
final_df.head()
```
